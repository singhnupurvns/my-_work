```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv("hiring (1).csv")
```

```python
df
```

```python
df["experience"] = df["experience"].fillna(0)
```

```python
df
```

```python
df["experience"].replace({"five":5,"two":2,"seven":7,"three":3,"ten":10,"eleven":11},inplace = True)
```

```python
df
```

```python
df["test_score(out of 10)"] =df["test_score(out of 10)"].fillna(df["test_score(out of 10)"].mean())
```

```python
df
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
x = df.drop(["salary($)"],axis = 1)
```

```python
y = df["salary($)"]
```

```python
le = model.fit(x,y)
```

```python
le
```

```python
prediction = le.predict(x)
```

```python
prediction1 = le.predict([[2,9,6]])
```

```python
prediction1
```

```python
prediction2 = le.predict([[12,10,10]])
```

```python
prediction2
```

```python
plt.figure(figsize = (14,10))
plt.scatter(df["experience"],df["salary($)"],color = "blue")
plt.title("Experience vs Salary")
plt.xlabel("Experience(years)")
plt.ylabel("Salary")

plt.figure(figsize = (14,10))
plt.scatter(df["test_score(out of 10)"],df["salary($)"],color = "red")
plt.title("Test_score vs Salary")
plt.xlabel("Test_score")
plt.ylabel("Salary")


plt.figure(figsize = (14,10))
plt.scatter(df["interview_score(out of 10)"],df["salary($)"],color = "black")
plt.title("Interview vs Salary")
plt.xlabel("Interview")
plt.ylabel("Salary")
plt.show()
```

```python
plt.figure(figsize = (10,6))
plt.plot(df["experience"],prediction,color = "red",linewidth = 2,label ="Predicted Salary")
plt.xlabel("Experience(years)")
plt.ylabel("Salary(&)")
plt.title("Linear Regression: Salary vs Experience")


plt.show()
```

```python
plt.figure(figsize = (10,6))
plt.plot(df["test_score(out of 10)"],prediction,color = "blue",linewidth = 2,label ="Predicted Salary")
plt.xlabel("Test_score(out of 10)")
plt.ylabel("Salary(&)")
plt.title("Linear Regression: Salary vs Test_score ")


plt.show()
```
